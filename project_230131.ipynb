```python
import numpy as np 
import pandas as pd
from pandas import DataFrame
```

```python
from google.colab import drive
drive.mount('/content/drive')

PATH = '/content/drive/MyDrive/Colab Notebooks/project/data/'
df = pd.read_csv(PATH + 'seoul_bds.csv', encoding='euc-kr')
```

```python
df = df.drop(columns = ['접수연도', '지번구분', '지번구분명', '계약기간', '신규갱신여부', '계약갱신권사용여부', '종전 보증금', '종전 임대료'], axis=1)
df.head()
```

```python
df.rename(columns = {
    '자치구코드' : 'SGG_CD',
    '자치구명' : 'SGG_NM',
    '법정동코드' : 'BJDONG_CD',
    '법정동명' : 'BJDONG_NM',
    '본번' : 'BOBN',
    '부번' : 'BUBN',
    '층' : 'FLR_NO',
    '계약일' : 'CNTRCT_DE',
    '전월세 구분' : 'RENT_GBN',
    '임대면적(㎡)' : 'RENT_AREA',
    '보증금(만원)' : 'RENT_GTN',
    '임대료(만원)' : 'RENT_FEE',
    '건물명' : 'BLDG_NM',
    '건축년도' : 'BUILD_YEAR',
    '건물용도' : 'HOUSE_GBN_NM'
    }, inplace = True)
```

```python
df.head()
```

```python
df = df.drop(df[(df['HOUSE_GBN_NM']=='단독다가구') | (df['HOUSE_GBN_NM']=='연립다세대') | (df['CNTRCT_DE'] < 20220101)].index, axis=0)
df['HOUSE_GBN_NM'].value_counts()
df = df.dropna(axis=0)
df.info()
```

```python
df['CNTRCT_DE'] = pd.to_datetime(df['CNTRCT_DE'], format='%Y%m%d')
df.head()
```
